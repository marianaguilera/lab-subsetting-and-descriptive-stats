####  What is the average Purchase Price?

#### Select all the customers we have in Spain

```python
# your answer here

orders[orders.Country=="Spain"]['CustomerID'].unique()
```

#### How many customers do we have in Spain?
##### Hint : Use value_counts()

```python
# your answer here

orders[orders.Country=="Spain"]['CustomerID'].value_counts()
```

```python
len(orders[orders.Country=="Spain"]['CustomerID'].unique())
```

#### Select all the customers who have bought more than 50 items ?

```python
# your answer here
cust_may50=orders.groupby('CustomerID').agg({'Quantity':'sum'})
cust_may50[cust_may50.Quantity>50]
```

#### Select orders from Spain that are above 50 items

```python
# your answer here

ord_may50=orders.groupby(['InvoiceNo','Country']).agg({'Quantity':'sum'}).reset_index()
ord_may50[(ord_may50.Country=="Spain")&(ord_may50.Quantity>50)]
```

#### Select all free orders

```python
# your answer here

ord_free=orders.groupby(['InvoiceNo']).agg({'amount_spent':'sum'}).reset_index()
ord_free[ord_free.amount_spent==0]
```

#### Select all orders that are 'lunch bag'
#### Hint: Use string functions

```python
# your answer here
orders[orders.Description.str.contains('lunch bag')]
```

#### Select all orders that are made in 2011 and are 'lunch bag'  

```python
# your answer here

orders[orders.Description.str.contains('lunch bag')&(orders.year==2011)]
```

#### Show the frequency distribution of the amount spent in Spain.

```python
# your answer here

Amount_spain=orders[(orders.Country=="Spain")]
Amount_spain.amount_spent.hist()
```

#### Select all orders made in the month of August

```python
# your answer here
orders[(orders.month==8)]
```

#### Select how many orders are made by countries in the month of August
##### Hint: Use value_counts()

```python
# your answer here
orders_month=orders.groupby(["Country","month"]).agg({'InvoiceNo':'count'}).reset_index()
orders_month[(orders_month.month==8)]
```

#### What's the  average amount of money spent by country

```python
# your answer here
orders.groupby("Country").agg({'amount_spent':'mean'})
```

#### What's the most expensive item?

```python
# your answer here

max_price= max(orders['UnitPrice'])

orders[orders['UnitPrice'] == max_price]
```

#### What was the average amount spent per year ?

```python
# your answer here

orders.groupby("year").agg({'amount_spent':'mean'})
```
